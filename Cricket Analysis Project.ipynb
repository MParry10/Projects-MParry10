```python
#Import necesary libraries
import pandas as pd
import json
```

Transformación de archivos json a CSV para emplearlos en Power BI

```python
with open ("t20_json_files/t20_wc_match_results.json") as f:
    data= json.load(f)
data[0]["matchSummary"]                           #Array o lista de datos del archivo Json estaban en una unica posición
df_match = pd.DataFrame(data[0]["matchSummary"])  #Conversión de los datos "match Summary" a base de datos/ Data Frame
df_match.head()                                   #mostrar primero registros de la dataframe creada de los partidos
```

```python
df_match.shape   # Forma de la data frame ( 45 rows and 7 columns)
```

```python
#Renombrar Scorecard para usarlo como primary key con otras tablas y/o datos a utilizar.Scorecard funciona como un identifcación del partido("Match ID")
# df_match.rename({"scorecard": "match"}, axis= 1, inplace= True)
df_match.rename({"match": "match_id"}, axis= 1, inplace= True) # por error de codigo previo( line anterior) , se cambia de match a match_id 
#Visualizar el cambio con .head()
df_match.head()
```

```python
#Creación de Diccionario para ayudar a relacionar las tablas de match results con batting summary pues tienen en comun pais vs pais
# Pais VS Pais ( independiente del orden de los paises 1 vs 2 o 2 vs 1) debe ser igual al ID del partido

match_ids_dict={}

for index, row in df_match.iterrows():
    key1 = row["team1"] + " Vs " + row["team2"]
    key2 = row["team2"] + " Vs " + row["team1"]
    
    match_ids_dict[key1] = row["match_id"]
    match_ids_dict[key2] = row["match_id"]

match_ids_dict
```

**Transformación de Archivo de resumen de bateo (Batting summary) de .json a .csv**

```python
with open ("t20_json_files/t20_wc_batting_summary.json") as M:  #Este archivo Json tiene multiples entradas, luego la idea es convertir sus registros (records) en una lista
    dato = json.load(M)
    
    all_records = []   # lista u arreglo variable temporal usada para almacenar los datos en los batting summary de los partidos
    
    for rec in dato:
        all_records.extend(rec["battingSummary"])             # extender el arreglo o lista de batting summary pues cada registro de batting summary en Json es un arreglo ( crear un arreglo de arreglos)
    
all_records   # tras ejecutar el for, ahora todos estan en una unica lista/arreglo; de aqui se puede generar el dataframe
df_batting= pd.DataFrame(all_records) # creacion del dataframe/database
df_batting.head() #Visualización de primeros datos del DataFrame generado    
```

```python
#Ejemplo de como funciona extend.
a= [2,3]
b= [4,5]
a.extend(b)
print (a)
```

```python
# Convertir la columna de dismissal en out/not_out aplicandole la condición a cada registro de out o not out.
#Se crea una nueva columna de out/not_out
# Se asigna a esta columna valores segun la de dismissal segun se cumplan condicion de largo ( funcion len)

df_batting["out/not_out"] = df_batting.dismissal.apply( lambda x : "out" if len(x)> 0 else "not_out")
df_batting.head(11)
```

```python
#Al no requerirse ya la columna de "dismissal" esta se elimina en el dataframe actual de df_batting agregando el parametro True

#df_batting.drop(columns= ["dismissal"], inplace= True)
df_batting.head(50)
```

```python
#Al aparecer caracteres raros en los nombres de los bateadores; estos se eliminan en Batsman name para limpiar los datos
df_batting["batsmanName"] = df_batting["batsmanName"].apply(lambda x : x.replace("â€",""))
df_batting["batsmanName"] = df_batting["batsmanName"].apply(lambda x : x.replace("\xa0",""))
df_batting.head(10)
```

```python
#Se agrega nueva columna de match ID al df_batting dataframe con los match ID respectivos usando el diccionario creado anteriormente para el dataframe de match

df_batting["match_id"] = df_batting["match"].map( match_ids_dict)

df_batting.head()
```
