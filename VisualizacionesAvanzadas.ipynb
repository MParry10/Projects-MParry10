```python
#Coordenadas y Diagonales
g= sns.FacetGrid(peliculas, row="Genero", col="Año", hue="Genero")
kws= dict( s= 50, linewidth= 0.5, edgecolor="black")
g = g.map(plt.scatter , "RatingsCriticos", "RatingsAudiencia", **kws)
g.set(xlim = (0,100), ylim = (0,100)) #rango de los ejes coordenados X y Y

for ax in g.axes.flat:
    ax.plot((0,100),(0,100), c= "gray", ls= "--") #Añade diagonal a todo la malla g de graficos
g.add_legend()
```
